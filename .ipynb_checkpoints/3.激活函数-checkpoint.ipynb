```python
import numpy as np
import matplotlib.pylab as plt
```

## 一. 阶跃函数

```python
# 阶跃函数
def step_function(x):
    y = x > 0
    return y.astype(np.int16)
```

```python
x = np.array([-1.0, 1.0, 2.0])
y = x > 0
y
```

```python
y = y.astype(np.int16)
y
```

```python
# 绘制阶跃函数的图形
x = np.arange(-5.0, 5.0, 0.1) # 在-5.0到5.0的范围内，以0.1为单位，生成NumPY数组([-5.0, -4.9, ..., 4.9])
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)  # 指定y轴范围
plt.show()
```

## 二. sigmoid函数

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.array([-1.0, 1.0, 2.0])
sigmoid(x)
```

## 三. ReLU函数

```python
def relu(x):
    return np.maximum(0, x)
```

```python
# ReLU函数的图像
x = np.arange(-1.0, 1.1, 0.1) # 在-5.0到5.0的范围内，以0.1为单位，生成NumPY数组([-5.0, -4.9, ..., 4.9])
y = relu(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)  # 指定y轴范围
plt.show()
```

## 四. softmax 函数

```python
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

以上函数容易出现溢出的问题，因为输入可能会很大，exp(a) 很容易溢出，因此改为以下方式：

```python
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a - c) # 溢出对策，减去输入中最大的值
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
a = np.array([0.3, 2.9, 4.0])
y = softmax(a)
print(y)
print(np.sum(y))
```
